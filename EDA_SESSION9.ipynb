```python
log_data=np.log(data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Exponetial')
plt.show()
```

```python
plt.subplot(1,2,1).hist(data,bins=30,label='Exponetial')
plt.legend()
plt.subplot(2,2,1).hist(log_data,bins=30,label='Normal')
plt.legend()
```

```python
sq_data=np.sqrt(data)
plt.hist(sq_data,bins=30,label='Squre_Exponetial')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='yeo-johnson')
exp_data=data.reshape(-1,1)
trans_data=power.fit_transform(exp_data)
plt.hist(trans_data)
```

```python
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='box-cox')
box_data=data.reshape(-1,1)
cox_data=power.fit_transform(box_data)
plt.hist(cox_data)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1).hist(trans_data,bins=30,label='yeo-johnson')
plt.legend()
plt.subplot(1,2,2).hist(cox_data,bins=30,label='box-cox')
plt.legend()
plt.show()

```

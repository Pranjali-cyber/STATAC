***Difference between loc and iloc.Using loc function filter any 3 columns and iloc
function filter any 2 columns with the index from 200 to 300.***

***'loc' and 'iloc' both are functions provided by pandas for data selection and manipulation
     loc- loc is used to access a group of rows and columns by labels 
    loc will take directly row name and column name.
    iloc-  iloc is used to access a group of rows and columns by integer positions.
        in iloc you will use slicing position also.***

**Find the distribution of gender/sex column within the dataset using matplotlib and
seaborn.***

```python
sex_unique=titanic_df['sex'].unique()
sex_count=[]
for i in sex_unique:
    sex=(titanic_df['sex'])==i
    sex_val=len(titanic_df[sex])
    sex_count.append(sex_val)
sex_df=pd.DataFrame(zip(sex_unique,sex_count),columns=['sex','sex_count'])
sex_df
```

```python
plt.figure(figsize=(5,5)) # Change figure lay out
plt.bar('sex','sex_count',
        data=sex_df)  # plot
plt.title("distribution of gender/sex")     # title of the plot
plt.xlabel("sex")    # X-axis name
plt.ylabel("sex_count")        # y-axis name
plt.show() 
```

```python
cdf=titanic_df['sex'].value_counts()
keys=cdf.keys()
plt.figure(figsize=(5,3))
sns.countplot(data=titanic_df,x='sex',order=keys)
plt.title('distribution of gender/sex')
plt.xlabel('sex')
plt.ylabel('Count')
plt.show()           
```

***Plot pie chart of categorical columns***

```python
cdf=titanic_df['sex'].value_counts()
keys=cdf.keys()
values=cdf.values
keys,values
```

```python
plt.pie(values,
explode=[0.1,0.1],
labels=keys,
autopct="%0.2f%%",
startangle=180,
radius=1)
plt.show()
```
